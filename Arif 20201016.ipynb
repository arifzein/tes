```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('vgsales.csv')
df.head()
```

```python
#code here
df_sales = df.groupby(['Genre']).mean()[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']]
df_sales
```

```python
df_t = df_sales.T
df_t
```

```python
import matplotlib.pyplot as plt
x = np.arange(4)
fig,ax = plt.subplots()
fig.set_size_inches(20,12)

#ax.bar(x, df_sales['NA_Sales'], width=0.3, label='NA_Sales')

ax.bar(x,df_t['Action'],width=0.05, label='Action', color='red')
ax.bar(x+0.05,df_t['Adventure'],width=0.05, label='Adventure', color='blue')
ax.bar(x+0.1,df_t['Fighting'],width=0.05, label='Fighting', color='purple')
ax.bar(x+0.15,df_t['Misc'],width=0.05, label='Misc', color='black')
ax.bar(x+0.20,df_t['Platform'],width=0.05, label='Platform', color='yellow')
ax.bar(x+0.25,df_t['Puzzle'],width=0.05, label='Puzzle', color='green')
ax.bar(x+0.30,df_t['Racing'],width=0.05, label='Racing', color='pink')
ax.bar(x+0.35,df_t['Role-Playing'],width=0.05, label='Role-Playing', color='red')
ax.bar(x+0.40,df_t['Shooter'],width=0.05, label='Shooter', color='blue')
ax.bar(x+0.45,df_t['Simulation'],width=0.05, label='Simulation', color='purple')
ax.bar(x+0.50,df_t['Sports'],width=0.05, label='Sports', color='black')
ax.bar(x+0.55,df_t['Strategy'],width=0.05, label='Strategy', color='yellow')

#ax.set_xticklabels('NA_Sales','EU_Sales','JP_Sales','Other_Sales')
bar_width = 0.35
plt.xticks(x + bar_width, ('NA_Sales','EU_Sales','JP_Sales','Other_Sales'))
plt.title("Main Sales Video Games By Genre")
plt.xlabel('Region Sales')
plt.ylabel('Mean Sales')
plt.legend()
#plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x = np.arange(12)
fig,ax = plt.subplots()
fig.set_size_inches(20,12)

#ax.bar(x, df_sales['NA_Sales'], width=0.3, label='NA_Sales')

ax.bar(x,df_sales['NA_Sales'],width=0.2,label='NA Sales')
ax.bar(x+0.2,df_sales['EU_Sales'],width=0.2,label='EU Sales')
ax.bar(x+0.4,df_sales['JP_Sales'],width=0.2,label='JP Sales')
ax.bar(x+0.6,df_sales['Other_Sales'],width=0.2,label='Other Sales')

ax.set_xticklabels(df_sales.index)
plt.legend()
plt.show()
```

```python
#code here
fig, ax = plt.subplots()
fig.set_size_inches(12, 8)
ax.bar(df_sales.index, df_sales['NA_Sales'], label='NA_Sales')
ax.bar(df_sales.index, df_sales['NA_Sales'], bottom=df_sales['EU_Sales'], 
           label='EU_Sales')
ax.bar(df_sales.index, df_sales['NA_Sales'], 
       bottom=df_sales['NA_Sales']+df_sales['EU_Sales'], label='JP_Sales')
ax.bar(df_sales.index, df_sales['NA_Sales'], 
       bottom=df_sales['EU_Sales']+df_sales['JP_Sales']+df_sales['Other_Sales'], 
       label='Other_Sales')
plt.legend()
plt.show()
```
